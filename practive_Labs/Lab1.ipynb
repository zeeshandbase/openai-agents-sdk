```python
# ----------------------------
# Ollama local Agent + Runner
# ----------------------------
import requests

# ----------------------------
# Minimal Agent class
# ----------------------------
class Agent:
    def __init__(self, name, instructions, model="gemma3:1b"):
        self.name = name
        self.instructions = instructions
        self.model = model

# ----------------------------
# Minimal RunResult class
# ----------------------------
class RunResult:
    def __init__(self, final_output, last_agent, new_items, raw_responses):
        self.final_output = final_output
        self.last_agent = last_agent
        self.new_items = new_items
        self.raw_responses = raw_responses

# ----------------------------
# Minimal Runner class
# ----------------------------
class Runner:
    @staticmethod
    async def run(agent, prompt):
        url = f"http://localhost:11434/v1/completions"
        payload = {
            "model": agent.model,
            "prompt": f"{agent.instructions}\nUser: {prompt}\nAI:",
            "max_tokens": 200
        }
        response = requests.post(url, json=payload)
        response.raise_for_status()
        data = response.json()
        text = data["completion"] if "completion" in data else data["choices"][0]["text"]
        return RunResult(
            final_output=text,
            last_agent=agent,
            new_items=[],
            raw_responses=[text]
        )

# ----------------------------
# Create agent and run
# ----------------------------
agent = Agent(name="Analyst", instructions="You analyze topics briefly.")

# In Jupyter, you can use await directly
result = await Runner.run(agent, "Why is Python popular for AI?")

# ----------------------------
# Print results
# ----------------------------
print("=" * 50)
print(f"Final Output:  {result.final_output[:100]}...")  # Show first 100 chars
print(f"Last Agent:    {result.last_agent.name}")
print(f"New Items:     {len(result.new_items)}")
print(f"Raw Responses: {len(result.raw_responses)}")
print("=" * 50)
```
